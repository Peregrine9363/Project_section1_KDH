```python
# 데이터셋 불러오기 - H1N1 백신 관련 데이터
import pandas as pd

target = 'vacc_h1n1_f'
train = pd.merge(pd.read_csv('https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/vacc_flu/train.csv'), 
                 pd.read_csv('https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/vacc_flu/train_labels.csv')[target], left_index=True, right_index=True)
test = pd.read_csv('https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/vacc_flu/test.csv')
sample_submission = pd.read_csv('https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/vacc_flu/submission.csv')
```

```python
# 중복치 제거
def data_wrangling(df):
    print("중복된 데이터의 비율", df.duplicated().sum()/len(df))
    df.drop_duplicates(keep='first', inplace=True)
    print("중복치 제거 결과 : ", df.shape)
    
    return df

train = data_wrangling(train)

train.nunique() # cardinality가 높은 값(>30)은 state

# 높은 카디널리티 제거
def cardinality_del(df):

    # 높은 카디널리티를 가지는 특성을 제거합니다.
    selected_cols = df.select_dtypes(include=["number", "object"])
    labels = selected_cols.nunique()  # 특성별 카디널리티 리스트
    selected_features = labels[labels <= 30].index.tolist()  # 카디널리티가 30보다 작은 특성만 선택합니다.
    df = df[selected_features]

    return df

train = cardinality_del(train)
test = cardinality_del(test)

# feature engineering
# 불필요한 특성 제거
# doctor_recc_seasonal - 계절성 독감 백신 투여 권유 여부 : 타겟이 다름
# child_under_6_months - 6개월 미만 영아와 접촉 여부 : 영아는 타인과 접촉이 적음
# health_insurance - 건강보험 가입 여부 : 건강보험과 상관없이 모든 대상자는 백신을 접종했음
# opinion_seas_vacc_effective - 계절성 백신의 효과에 대한 응답자의 생각 : 타겟이 다름
# opinion_seas_risk - 계절성 백신의 위험에 대한 응답자의 생각 : 타겟이 다름
# opinion_seas_sick_from_vacc - 계절성 백신 투여 후 질병에 걸릴 수도 있는 위험에 대한 응답자의 생각 : 타겟이 다름
# rent_own_r - 응답자의 거주 현황 : 소득과 근로 정보가 있어서 자택 소유 여부는 상관이 없음
# n_adult_r - 거주 구성원 중 어른 인원 수 : n_people_r(거주 구성원의 총 인원 수)로 통합 가능
# household_children - 거주 구성원 중 어린이 인원 수 : 위와 동일
# hhs_region - HHS 감시 지역 번호 : HHS 감시 지역과 백신 반응은 상관관계가 낮음
# state - 거주하는 주(州) : 인구조사 거주 지역 자료로 충분하다고 판단 (카디널리티가 높아서 앞에서 제거했음)
# employment_industry - 응답자가 종사하는 산업 구분 : 결측치가 많고 상대적으로 관계가 적다고 판단
# employment_occupation - 응답자의 직업 유형 (짧은 임의의 문자열로 표시) : 위와 동일

check = train[(train.n_adult_r[:] + train.household_children[:] != train.n_people_r[:])] # 값 다른 이유 확인
train.n_people_r[:] = train.n_adult_r[:] + train.household_children[:] # 거주 구성원의 총 인원 수로 통합

def feature_eng(df):
    cols = ['doctor_recc_seasonal', 'child_under_6_months', 'health_insurance', 'opinion_seas_vacc_effective',
            'opinion_seas_risk', 'opinion_seas_sick_from_vacc', 'rent_own_r', 'n_adult_r',
            'household_children', 'hhs_region', 'employment_industry', 'employment_occupation']
    df.drop(cols, axis=1, inplace=True)
    
    print(df.shape)
    
    return df

feature_eng(train)
feature_eng(test)

# 데이터 셋 나누기
from sklearn.model_selection import train_test_split

features = test.columns

train, val = train_test_split(train, train_size=0.80, test_size=0.20, stratify=train[target], random_state=2)

print(train.shape, val.shape, test.shape)

X_train = train[features]
y_train = train[target]
X_val = val[features]
y_val = val[target]
X_test = test[features]
```

##### **Q1. early stopping 적용 후 best iteration을 제출해 주세요.**

```python
from category_encoders import OrdinalEncoder
from sklearn.impute import SimpleImputer
from xgboost import XGBClassifier
from sklearn.pipeline import make_pipeline

# Early Stopping 미적용
pipe = make_pipeline(
    OrdinalEncoder(),
    SimpleImputer(strategy="median"),
    XGBClassifier(
        objective="binary:logistic",
        eval_metric="error",  # error = 1 - accuracy 지표를 사용해 평가합니다.
        n_estimators=1000,
        random_state=2,
        n_jobs=-1,
        max_depth=7,
        learning_rate=0.1,
        reg_lambda=1,
    ),
)
print(pipe)

# 모델을 학습합니다.
pipe.fit(X_train, y_train)

from sklearn.metrics import classification_report

# train 학습, 검증셋 정확도
print("검증 정확도", pipe.score(X_val, y_val))

y_pred = pipe.predict(X_val)
print(classification_report(y_val, y_pred))

# Early Stopping 적용
encoder = OrdinalEncoder()
X_train_encoded = encoder.fit_transform(X_train)  # 학습데이터
X_val_encoded = encoder.transform(X_val)  # 검증데이터
X_test_encoded = encoder.transform(X_test) # 테스트데이터

model = XGBClassifier(
    objective="binary:logistic",
    eval_metric="error",  # error = 1 - accuracy 지표를 사용해 평가합니다.
    n_estimators=1000,  # 매우 큰 값으로 설정했지만, 성능 향상이 없으면 자동으로 종료합니다.
    random_state=2,
    n_jobs=-1,
    max_depth=7,
    learning_rate=0.1,
    reg_lambda=1,
)

watchlist = [(X_train_encoded, y_train), (X_val_encoded, y_val)]

model.fit(
    X_train_encoded,
    y_train,
    eval_set=watchlist,
    early_stopping_rounds=200,  # 200 rounds 동안 성능 개선이 없으면 학습을 중지합니다.
)

print(model.best_iteration)

# train 학습, 검증셋 정확도
print("검증 정확도", model.score(X_val_encoded, y_val))

y_pred = model.predict(X_val_encoded)
print(classification_report(y_val, y_pred))
```

## **Part 3. 모델 성능 향상**
- Xgboost의 하이퍼파라미터를 조정하여 성능을 향상시키세요.
  - F1-score 0.61을 달성해 보세요!

```python
from sklearn.metrics import accuracy_score, f1_score
import numpy as np

# XGBoost early_stopping_rounds 적용 함수
watchlist = [(X_train_encoded, y_train), (X_val_encoded, y_val)]

def XGBoost_esr(lr, ne, md, mcw):
    model = XGBClassifier(
        objective="binary:logistic",
        eval_metric="error",  # error = 1 - accuracy 지표를 사용해 평가합니다.
        random_state=2,
        n_jobs=-1,
        reg_lambda=1,
        learning_rate=lr,
        n_estimators=ne,
        max_depth=md,
        min_child_weight=mcw
    )

    model.fit(
        X_train_encoded,
        y_train,
        eval_set=watchlist,
        early_stopping_rounds=200,  # 200 rounds 동안 성능 개선이 없으면 학습을 중지합니다.
    )
    
    y_pred = model.predict(X_val_encoded)
    
    f1 = f1_score(y_val, y_pred)
    
    return f1
```

```python
lr = 0.3 # learning_rate [ 기본값 : 0.3 ]
ne = 100 # n_estimators [ 기본값 : 100 ]
md = 6 # max_depth [ 기본값 : 6 ]
mcw = 1 # min_child_weight [ 기본값 : 1 ]

XGBoost_esr(lr, ne, md, mcw)
```

```python
# 최적 하이퍼파라미터 값 찾기
def find_best_hp(lr, ne, md, mcw):
    f1 = 0
    num = 0
    
    for a in lr:
        for b in ne:
            for c in md:
                for d in mcw:
                    num = num + 1
                    count = ((num) / (len(lr) * len(ne) * len(md) * len(mcw))) * 100
                    print("계산 완료 (%) : ", round(count, 3))
                    
                    if f1 < XGBoost_esr(a, b, c, d):
                        f1 = XGBoost_esr(a, b, c, d)
                        lr_val = a
                        ne_val = b
                        md_val = c
                        mcw_val = d
                        
    print('learning_rate : ', lr_val)
    print('n_estimators : ', ne_val)
    print('max_depth : ', md_val)
    print('min_child_weight : ', mcw_val)
    print("f1 score : ", round(f1, 3))
```

```python
lr = [0.3, 0.5, 0.7] # learning_rate [ 기본값 : 0.3 ] 높을수록 과적합
ne = [1000] # n_estimators [ 기본값 : 100 ] 높을수록 과적합 방지
md = [7, 8, 9, 10] # max_depth [ 기본값 : 6 ] 높을수록 과적합
mcw = [1] # min_child_weight [ 기본값 : 1 ] 높을수록 과적합 방지
```

```python
find_best_hp(lr, ne, md, mcw)
```

```python
# ë°ì´í„°ì…‹ ë¶ˆëŸ¬ì˜¤ê¸° - H1N1 ë°±ì‹  ê´€ë ¨ ë°ì´í„°
import pandas as pd

target = 'vacc_h1n1_f'
train = pd.merge(pd.read_csv('https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/vacc_flu/train.csv'), 
                 pd.read_csv('https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/vacc_flu/train_labels.csv')[target], left_index=True, right_index=True)
test = pd.read_csv('https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/vacc_flu/test.csv')
sample_submission = pd.read_csv('https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/vacc_flu/submission.csv')
```

```python
# ì¤‘ë³µì¹˜ ì œê±°
def data_wrangling(df):
    print("ì¤‘ë³µëœ ë°ì´í„°ì˜ ë¹„ìœ¨", df.duplicated().sum()/len(df))
    df.drop_duplicates(keep='first', inplace=True)
    print("ì¤‘ë³µì¹˜ ì œê±° ê²°ê³¼ : ", df.shape)
    
    return df

train = data_wrangling(train)

train.nunique() # cardinalityê°€ ë†’ì€ ê°’(>30)ì€ state

# ë†’ì€ ì¹´ë””ë„ë¦¬í‹° ì œê±°
def cardinality_del(df):

    # ë†’ì€ ì¹´ë””ë„ë¦¬í‹°ë¥¼ ê°€ì§€ëŠ” íŠ¹ì„±ì„ ì œê±°í•©ë‹ˆë‹¤.
    selected_cols = df.select_dtypes(include=["number", "object"])
    labels = selected_cols.nunique()  # íŠ¹ì„±ë³„ ì¹´ë””ë„ë¦¬í‹° ë¦¬ìŠ¤íŠ¸
    selected_features = labels[labels <= 30].index.tolist()  # ì¹´ë””ë„ë¦¬í‹°ê°€ 30ë³´ë‹¤ ì‘ì€ íŠ¹ì„±ë§Œ ì„ íƒí•©ë‹ˆë‹¤.
    df = df[selected_features]

    return df

train = cardinality_del(train)
test = cardinality_del(test)

# feature engineering
# ë¶ˆí•„ìš”í•œ íŠ¹ì„± ì œê±°
# doctor_recc_seasonal - ê³„ì ˆì„± ë…ê° ë°±ì‹  íˆ¬ì—¬ ê¶Œìœ  ì—¬ë¶€ : íƒ€ê²Ÿì´ ë‹¤ë¦„
# child_under_6_months - 6ê°œì›” ë¯¸ë§Œ ì˜ì•„ì™€ ì ‘ì´‰ ì—¬ë¶€ : ì˜ì•„ëŠ” íƒ€ì¸ê³¼ ì ‘ì´‰ì´ ì ìŒ
# health_insurance - ê±´ê°•ë³´í—˜ ê°€ì… ì—¬ë¶€ : ê±´ê°•ë³´í—˜ê³¼ ìƒê´€ì—†ì´ ëª¨ë“  ëŒ€ìƒìëŠ” ë°±ì‹ ì„ ì ‘ì¢…í–ˆìŒ
# opinion_seas_vacc_effective - ê³„ì ˆì„± ë°±ì‹ ì˜ íš¨ê³¼ì— ëŒ€í•œ ì‘ë‹µìì˜ ìƒê° : íƒ€ê²Ÿì´ ë‹¤ë¦„
# opinion_seas_risk - ê³„ì ˆì„± ë°±ì‹ ì˜ ìœ„í—˜ì— ëŒ€í•œ ì‘ë‹µìì˜ ìƒê° : íƒ€ê²Ÿì´ ë‹¤ë¦„
# opinion_seas_sick_from_vacc - ê³„ì ˆì„± ë°±ì‹  íˆ¬ì—¬ í›„ ì§ˆë³‘ì— ê±¸ë¦´ ìˆ˜ë„ ìˆëŠ” ìœ„í—˜ì— ëŒ€í•œ ì‘ë‹µìì˜ ìƒê° : íƒ€ê²Ÿì´ ë‹¤ë¦„
# rent_own_r - ì‘ë‹µìì˜ ê±°ì£¼ í˜„í™© : ì†Œë“ê³¼ ê·¼ë¡œ ì •ë³´ê°€ ìˆì–´ì„œ ìíƒ ì†Œìœ  ì—¬ë¶€ëŠ” ìƒê´€ì´ ì—†ìŒ
# n_adult_r - ê±°ì£¼ êµ¬ì„±ì› ì¤‘ ì–´ë¥¸ ì¸ì› ìˆ˜ : n_people_r(ê±°ì£¼ êµ¬ì„±ì›ì˜ ì´ ì¸ì› ìˆ˜)ë¡œ í†µí•© ê°€ëŠ¥
# household_children - ê±°ì£¼ êµ¬ì„±ì› ì¤‘ ì–´ë¦°ì´ ì¸ì› ìˆ˜ : ìœ„ì™€ ë™ì¼
# hhs_region - HHS ê°ì‹œ ì§€ì—­ ë²ˆí˜¸ : HHS ê°ì‹œ ì§€ì—­ê³¼ ë°±ì‹  ë°˜ì‘ì€ ìƒê´€ê´€ê³„ê°€ ë‚®ìŒ
# state - ê±°ì£¼í•˜ëŠ” ì£¼(å·) : ì¸êµ¬ì¡°ì‚¬ ê±°ì£¼ ì§€ì—­ ìë£Œë¡œ ì¶©ë¶„í•˜ë‹¤ê³  íŒë‹¨ (ì¹´ë””ë„ë¦¬í‹°ê°€ ë†’ì•„ì„œ ì•ì—ì„œ ì œê±°í–ˆìŒ)
# employment_industry - ì‘ë‹µìê°€ ì¢…ì‚¬í•˜ëŠ” ì‚°ì—… êµ¬ë¶„ : ê²°ì¸¡ì¹˜ê°€ ë§ê³  ìƒëŒ€ì ìœ¼ë¡œ ê´€ê³„ê°€ ì ë‹¤ê³  íŒë‹¨
# employment_occupation - ì‘ë‹µìì˜ ì§ì—… ìœ í˜• (ì§§ì€ ì„ì˜ì˜ ë¬¸ìì—´ë¡œ í‘œì‹œ) : ìœ„ì™€ ë™ì¼

check = train[(train.n_adult_r[:] + train.household_children[:] != train.n_people_r[:])] # ê°’ ë‹¤ë¥¸ ì´ìœ  í™•ì¸
train.n_people_r[:] = train.n_adult_r[:] + train.household_children[:] # ê±°ì£¼ êµ¬ì„±ì›ì˜ ì´ ì¸ì› ìˆ˜ë¡œ í†µí•©

def feature_eng(df):
    cols = ['doctor_recc_seasonal', 'child_under_6_months', 'health_insurance', 'opinion_seas_vacc_effective',
            'opinion_seas_risk', 'opinion_seas_sick_from_vacc', 'rent_own_r', 'n_adult_r',
            'household_children', 'hhs_region', 'employment_industry', 'employment_occupation']
    df.drop(cols, axis=1, inplace=True)
    
    print(df.shape)
    
    return df

feature_eng(train)
feature_eng(test)

# ë°ì´í„° ì…‹ ë‚˜ëˆ„ê¸°
from sklearn.model_selection import train_test_split

features = test.columns

train, val = train_test_split(train, train_size=0.80, test_size=0.20, stratify=train[target], random_state=2)

print(train.shape, val.shape, test.shape)

X_train = train[features]
y_train = train[target]
X_val = val[features]
y_val = val[target]
X_test = test[features]
```

## **Part 2. Xgboost**

### **2-1. ëª¨ë¸ í•™ìŠµ**
- Xgboostë¥¼ í•™ìŠµì‹œí‚¤ì„¸ìš”.
  - `random_state=2`ë¡œ ì„¤ì •í•©ë‹ˆë‹¤. 
  - `n_estimators=1000`, `early_stopping_rounds=200`ìœ¼ë¡œ ì„¤ì •í•œ í›„ best iterationì„ êµ¬í•˜ì„¸ìš”. 
- ê²€ì¦ ì„¸íŠ¸ ìŠ¤ì½”ì–´ë¥¼ í™•ì¸í•˜ì„¸ìš”. 
  - í‰ê°€ì§€í‘œëŠ” F1-scoreì…ë‹ˆë‹¤.

##### **Q1. early stopping ì ìš© í›„ best iterationì„ ì œì¶œí•´ ì£¼ì„¸ìš”.**

```python
from category_encoders import OrdinalEncoder
from sklearn.impute import SimpleImputer
from xgboost import XGBClassifier
from sklearn.pipeline import make_pipeline

# Early Stopping ë¯¸ì ìš©
pipe = make_pipeline(
    OrdinalEncoder(),
    SimpleImputer(strategy="median"),
    XGBClassifier(
        objective="binary:logistic",
        eval_metric="error",  # error = 1 - accuracy ì§€í‘œë¥¼ ì‚¬ìš©í•´ í‰ê°€í•©ë‹ˆë‹¤.
        n_estimators=1000,
        random_state=2,
        n_jobs=-1,
        max_depth=7,
        learning_rate=0.1,
        reg_lambda=1,
    ),
)
print(pipe)

# ëª¨ë¸ì„ í•™ìŠµí•©ë‹ˆë‹¤.
pipe.fit(X_train, y_train)

from sklearn.metrics import classification_report

# train í•™ìŠµ, ê²€ì¦ì…‹ ì •í™•ë„
print("ê²€ì¦ ì •í™•ë„", pipe.score(X_val, y_val))

y_pred = pipe.predict(X_val)
print(classification_report(y_val, y_pred))

# Early Stopping ì ìš©
encoder = OrdinalEncoder()
X_train_encoded = encoder.fit_transform(X_train)  # í•™ìŠµë°ì´í„°
X_val_encoded = encoder.transform(X_val)  # ê²€ì¦ë°ì´í„°
X_test_encoded = encoder.transform(X_test) # í…ŒìŠ¤íŠ¸ë°ì´í„°

model = XGBClassifier(
    objective="binary:logistic",
    eval_metric="error",  # error = 1 - accuracy ì§€í‘œë¥¼ ì‚¬ìš©í•´ í‰ê°€í•©ë‹ˆë‹¤.
    n_estimators=1000,  # ë§¤ìš° í° ê°’ìœ¼ë¡œ ì„¤ì •í–ˆì§€ë§Œ, ì„±ëŠ¥ í–¥ìƒì´ ì—†ìœ¼ë©´ ìë™ìœ¼ë¡œ ì¢…ë£Œí•©ë‹ˆë‹¤.
    random_state=2,
    n_jobs=-1,
    max_depth=7,
    learning_rate=0.1,
    reg_lambda=1,
)

watchlist = [(X_train_encoded, y_train), (X_val_encoded, y_val)]

model.fit(
    X_train_encoded,
    y_train,
    eval_set=watchlist,
    early_stopping_rounds=200,  # 200 rounds ë™ì•ˆ ì„±ëŠ¥ ê°œì„ ì´ ì—†ìœ¼ë©´ í•™ìŠµì„ ì¤‘ì§€í•©ë‹ˆë‹¤.
)

print(model.best_iteration)

# train í•™ìŠµ, ê²€ì¦ì…‹ ì •í™•ë„
print("ê²€ì¦ ì •í™•ë„", model.score(X_val_encoded, y_val))

y_pred = model.predict(X_val_encoded)
print(classification_report(y_val, y_pred))
```

```python
from sklearn.metrics import accuracy_score, f1_score
import numpy as np

# XGBoost early_stopping_rounds ì ìš© í•¨ìˆ˜
watchlist = [(X_train_encoded, y_train), (X_val_encoded, y_val)]

def XGBoost_esr(lr, ne, md, mcw):
    model = XGBClassifier(
        objective="binary:logistic",
        eval_metric="error",  # error = 1 - accuracy ì§€í‘œë¥¼ ì‚¬ìš©í•´ í‰ê°€í•©ë‹ˆë‹¤.
        random_state=2,
        n_jobs=-1,
        reg_lambda=1,
        learning_rate=lr,
        n_estimators=ne,
        max_depth=md,
        min_child_weight=mcw
    )

    model.fit(
        X_train_encoded,
        y_train,
        eval_set=watchlist,
        early_stopping_rounds=200,  # 200 rounds ë™ì•ˆ ì„±ëŠ¥ ê°œì„ ì´ ì—†ìœ¼ë©´ í•™ìŠµì„ ì¤‘ì§€í•©ë‹ˆë‹¤.
    )
    
    y_pred = model.predict(X_val_encoded)
    
    f1 = f1_score(y_val, y_pred)
    
    return f1
```
